```python
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content']=df['Item Fat Content'].replace({'LF':'Low Fat',
                                                       'low fat':'Low Fat',
                                                       'reg':'Regular'})
```

# **Bussiness Requirements**


###  **KPI'S Requirements**


```python
#Total Sales
Total_Sales=df['Sales'].sum()
#Average Sales
Average_Sales=df['Sales'].mean()
#number of items sold
number_of_items_sold=df['Sales'].count()
#Average Ratings
Average_Ratings=df['Rating'].mean()
#Display
print(f"Total Sales: $ {Total_Sales:,.1f}")
print(f"Average Sales: $ {Average_Sales:,.1f}")
print(f"number of items sold: {number_of_items_sold:,.1f}")
print(f"Average Ratings {Average_Ratings:,.1f}")
```

# **chart  Requirements**


###  **Total Sales by fat content**


```python
sales_by_fat=df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(sales_by_fat,labels= sales_by_fat.index,
        autopct='%.1f%%',
        startangle=90)
plt.title('Total Sales by fat content')
plt.axis('equal')
plt.show()
```

###  **Total Sales by item type**


```python
sales_by_type=df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
plt.bar(sales_by_type.index,sales_by_type.values,color='pink')
plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by item type')
plt.tight_layout()
plt.show()
```

###  **Fat content by outlet for total sales**


```python
grouped= df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
grouped=grouped[['Regular' ,'Low Fat']]
ax=grouped.plot(kind='bar',figsize=(8,5),title='outlet tier by item fat content')
plt.xlabel('outlet location tier')
plt.ylabel('total sales')
plt.legend(title='Item Fat Content')
#plt.grid()
plt.tight_layout()
plt.show()
```

### **Sales by outlet location**


```python
sales_by_outlet_loc=df.groupby('Outlet Location Type')['Sales'].sum()
plt.figure(figsize=(8,3))
plt.barh(sales_by_outlet_loc.index,sales_by_outlet_loc.values)
plt.xlabel('Outlet Location Type')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

### **Sales by outlet Establishment**


```python
sales_by_outlet_year=df.groupby('Outlet Establishment Year')['Sales'].sum()
plt.figure(figsize=(9,4))
plt.plot(sales_by_outlet_year,marker='s',linestyle='-',color='red',markerfacecolor='blue')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Sales')
plt.title('sales by outlet Establishment')
plt.tight_layout()
plt.show()
```

### **Sales by outlet size**


```python
sales_by_outlet_size=df.groupby('Outlet Size')['Sales'].sum()
plt.pie(sales_by_outlet_size,
        labels=sales_by_outlet_size.index,
        autopct='%.1f%%',
        startangle=90,
        wedgeprops={'width':0.5} )  # controls the size of the hole
plt.axis('equal')
plt.title('sales by outlet size')
plt.show()
```
